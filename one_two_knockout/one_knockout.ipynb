```python
import cobra
import time
import csv
from tqdm import tqdm
```

```python
import os
directory = '/data1/xpgeng/cross_pathogen/GEM_model'
files = os.listdir(directory)
print(files)
```

#### iML1515

```python
model = cobra.io.read_sbml_model('/data1/xpgeng/cross_pathogen/GEM_model/iML1515.xml')
```

```python
model.summary()
```

```python
with open('iML1515_1KO_growth', mode='a', newline='') as csvfile:
    writer = csv.writer(csvfile)
    firstrow = ['gene, label']
    writer.writerow(firstrow)  
    geneid = []
    for i in model.genes:
        geneid.append(i.id) 

    state_0   = cobra.flux_analysis.pfba(model)      
    growth_0    = round(state_0['BIOMASS_Ec_iML1515_core_75p37M'], 3)

    for i in tqdm(geneid):   
        with model:                
            try:            
                model.genes.get_by_id(i).knock_out()
                state_1   = cobra.flux_analysis.pfba(model)
                growth_1    = round(state_1['BIOMASS_Ec_iML1515_core_75p37M'], 3)
                if growth_1 < 0.1*growth_0:
                    
                    writer.writerow([i, growth_0, growth_1, 1]) 
                else:
                    writer.writerow([i, growth_0, growth_1, 0])
            except:
                writer.writerow([i, 'fail'])
```
